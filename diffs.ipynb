```python
import os
import difflib

root_src_dir = './source_code/'
root_old_dir = './source_code_jl/'
root_py_dir = './source_code_py/'

diffObj = difflib.HtmlDiff(linejunk=difflib.IS_LINE_JUNK, charjunk=difflib.IS_CHARACTER_JUNK)


with open("diffs.html", "w") as g:

    g.write("<html>\n")
    g.write("<head>\n")
    g.write("<meta charset=\"utf-8\">\n")
    g.write("<style>\n")
    g.write("html, body {font-size: 10px}\n")
    g.write("h2 {display: inline-block;}\n")

    g.write("""
        table.diff {font-family:Courier; border:medium; font-size:12px;}
        .diff_header {background-color:#e0e0e0}
        td.diff_header {text-align:right}
        .diff_next {background-color:#c0c0c0}
        .diff_add {background-color:#aaffaa}
        .diff_chg {background-color:#ffff77}
        .diff_sub {background-color:#ffaaaa}
""")
    g.write("</style>\n")
    g.write("</head>\n")
    g.write("<body>\n")

    g.write("""<table class="diff" summary="Legends">
        <tr> <th colspan="2"> Legends </th> </tr>
        <tr> <td> <table border="" summary="Colors">
                      <tr><th> Colors </th> </tr>
                      <tr><td class="diff_add">&nbsp;Added&nbsp;</td></tr>
                      <tr><td class="diff_chg">Changed</td> </tr>
                      <tr><td class="diff_sub">Deleted</td> </tr>
                  </table></td>
             <td> <table border="" summary="Links">
                      <tr><th colspan="2"> Links </th> </tr>
                      <tr><td>(f)irst change</td> </tr>
                      <tr><td>(n)ext change</td> </tr>
                      <tr><td>(t)op</td> </tr>
                  </table></td> </tr>
    </table>""")

    for file in os.listdir(root_src_dir):
            target = os.path.join(root_old_dir, file)

            pyFile = file.replace(".jl", ".py")

            if os.path.isfile(os.path.join(root_py_dir, pyFile)):
                flag = f" - Python file exists"
                color = "#5b7c99"
            else:
                flag = f" - Python file {pyFile} does not exist"
                color = "#fde74c"

            if os.path.isfile(target):
                    with open(os.path.join(root_src_dir, file), "r") as f:
                        src = f.readlines()
                    with open(target, "r") as f:
                        old = f.readlines()

                    if len(list(difflib.unified_diff(old, src))) == 0:
                        g.write(f"<h1>{file}</h1>\n")
                        g.write(f"<h2><span style=\"background-color:#ffeab8;\"> is unchanged</span></h2>\n")
                        g.write(f"<h2><span style=\"background-color:{color};\">{flag}</span></h2>\n")
                        continue
                
                    g.write(f"<h1>{file}</h1>\n")
                    g.write(f"<h2><span style=\"background-color:#ff700a;\"> is changed</span></h2>\n")
                    g.write(f"<h2><span style=\"background-color:{color};\">{flag}</span></h2>\n")

                    diff = diffObj.make_table(fromlines=old, tolines=src, fromdesc="old public code", todesc="new private (jstac) code", context=True, numlines=2)


                    g.write(diff)
            else:
                g.write(f"<h1>{file}</h1>\n")
                g.write(f"<h2><span style=\"background-color:#a6deae;\">new</span></h2>\n")
                g.write(f"<h2><span style=\"background-color:{color};\">{flag}</span></h2>\n")

    g.write("</body>\n")
    g.write("</html>\n")
```
